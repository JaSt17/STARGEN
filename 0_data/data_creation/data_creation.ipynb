This Notebooks transforms the 4 given excel sheets into one usable excel sheet

```python
import pandas as pd
import openpyxl
```

```python
# read in the Ghalichi dataset
Ghalichi_30GPs = pd.read_excel('Ghalichi_2024_30GPs.xlsx', header=None)
Ghalichi_annots = pd.read_excel('Ghalichi_2024.xlsx')
```

```python
# Set the column names that we want to keep
keep_cols = ['Poseidon_ID', 'Publication', 'Date_C14_Uncal_BP', 'Site', 'Country', 'Latitude', 'Longitude']
# Only keep the columns we need
Ghalichi_annots = Ghalichi_annots[keep_cols]
# Change the publication column to Ghalichi_2024
Ghalichi_annots['Publication'] = 'Ghalichi_2024'
# cahnge the Site column to Locality and the Country column to political Entity
Ghalichi_annots.rename(columns={'Poseidon_ID':'Genetic ID','Date_C14_Uncal_BP':'Date BP', 'Site':'Locality', 'Country':'Political_Entity', 'Latitude':'Lat.', "Longitude":"Long."}, inplace=True)
Ghalichi_annots.head()
```

```python
# set a temporary header to the Ghalichi_30GPs dataframe
cols = [i for i in range(1, len(Ghalichi_30GPs.columns)+1)]
cols[1] = 'Genetic ID'
# set the columns to keep in the Ghalichi_30GPs dataframe so we only keep the genetic ID and the admixture proportions
keep_cols = [i for i in range(6, len(Ghalichi_30GPs.columns)+1)]
keep_cols[0] = 'Genetic ID'
Ghalichi_30GPs.columns = cols
Ghalichi_30GPs = Ghalichi_30GPs[keep_cols]
Ghalichi_30GPs.head()
```

```python
# Merge the two dataframes on the Genetic ID column
merged_Ghalichi = pd.merge(Ghalichi_annots, Ghalichi_30GPs, on='Genetic ID')
merged_Ghalichi.head()
```

```python
# load the AADR_V62.0 datasets
AADR_V62_30GPs = pd.read_excel('Reich dataset V62.0_30GPs.xlsx', header=None)
AADR_V62_annots = pd.read_excel('Reich dataset V62.0.xlsx')
```

```python
# Set the column names that we want to keep
keep_cols = ['Genetic ID (suffixes: ".DG" is a high coverage shotgun genome with diploid genotype calls, ".AG" is shotgun data with each position in the genome represented by a randomly chosen sequence, ".HO" is Affymetrix Human Origins genotype data)',
                'Publication abbreviation',
                'Date mean in BP in years before 1950 CE [OxCal mu for a direct radiocarbon date, and average of range for a contextual date]',
                'Locality', 'Political Entity', 'Lat.', 'Long.']
# Only keep the columns we need
AADR_V62_annots = AADR_V62_annots[keep_cols]
AADR_V62_annots.rename(columns={'Genetic ID (suffixes: ".DG" is a high coverage shotgun genome with diploid genotype calls, ".AG" is shotgun data with each position in the genome represented by a randomly chosen sequence, ".HO" is Affymetrix Human Origins genotype data)':'Genetic ID',
                                'Publication abbreviation':'Publication',
                                'Date mean in BP in years before 1950 CE [OxCal mu for a direct radiocarbon date, and average of range for a contextual date]':'Date BP'}, inplace=True)
AADR_V62_annots.head()
```

```python
# set temp header to the AADR_V62_30GPs dataframe
cols = [i for i in range(1, len(AADR_V62_30GPs.columns)+1)]
cols[1] = 'Genetic ID'
AADR_V62_30GPs.columns = cols
# drop the first column
AADR_V62_30GPs.drop(columns=[1], inplace=True)
```

```python
# merge the two dataframes on the Genetic ID column
merged_AADR_V62 = pd.merge(AADR_V62_annots, AADR_V62_30GPs, on='Genetic ID')
merged_AADR_V62.head()
```

```python
# change the names of the admixture proportions columns in both dataframes
col_names = ['Genetic ID', 'Publication',
            'Date_BP', 'Locality', 'Political_Entity', 'Lat.', 'Long.',
            'As_Ru_3K_0.9K_BCE', 'As_Ru_5.8K_3.6K_BCE', 'As_Ru_37.4K_3K_BCE',
            'As_Ru_E_0.1K_1.4K_CE', 'C_Af_6K_1K_BCE', 'C_NW_As_2.2K_1.6K_BCE',
            'C_N_As_BCE', 'C_N_As_CE', 'E_Af_2.5KBCE_1.6KCE', 'E_As_6KBCE_0.25KCE',
            'E_C_Eu_6K_2K_BCE', 'E_Eu_9K_3K_BCE', 'E_Eu_CE', 'E_Eu_Po_3K_2.3K_BCE',
            'NN_Am_1K_2.0K_CE', 'N_Am_11.5KBCE_1.2KCE', 'N_Eu_0.7K_1.7K_CE',
            'Oceania_1KBCE_2KCE', 'SE_As_6KBCE_0.8KCE', 'SW_As_8K_1.1K_BCE',
            'SW_C_As_9K_0.8K_BCE', 'S_Am_CE', 'S_Eu_5.4K_1K_BCE', 'S_Eu_16K_5K_BCE',
            'S_Eu_CE', 'S_M_Am_3.5KBCE_1.7KCE', 'W_Eu_5.5K_2.5K_BCE',
            'W_Eu_30K_5K_BCE', 'W_Eu_38K_BCE', 'W_Eu_CE']
# change the column names in the merged dataframes
merged_Ghalichi.columns = col_names
merged_AADR_V62.columns = col_names

# merge the two dataframes
merged = pd.concat([merged_Ghalichi, merged_AADR_V62])
merged.shape
```

```python
# write the merged dataframe to an excel file
merged.to_excel('aDNA_30GPs.xlsx', index=False)
```
